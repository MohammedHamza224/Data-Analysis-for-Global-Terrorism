```python
# First cheaking for Null Values in case found in data ..!
df.isnull().sum().sort_values(ascending=False)
```

```python
# Return percentage of missing values in each column 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}  -  {}%'.format(col, round(pct_missing*100)))
```

```python
# Drop columns that contain 30% or more  missing values 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    pct_missing =  round(pct_missing*100)
    if pct_missing >= 30 :
      df.drop([col], axis = 1, inplace = True)
      print('Droping {} Column as it contain - {}% null values'.format(col, pct_missing))
```

```python
# Cheaking for shape of data after null values clean
df.shape
```

```python
# Return percentage of missing values in each column  
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}  -  {}%'.format(col, round(pct_missing*100)))
```

```python
# -- Handle missing values:
# - Fill missing values in categorical columns with the most frequent value (mode).
# - Fill missing values in numerical columns with the column mean.
for column in df.columns:
    if df[column].dtype == 'object':  # For categorical columns
        df[column] = df[column].fillna(df[column].mode()[0])
    elif df[column].dtype in ['int64', 'float64']:  # For numerical columns
        df[column] = df[column].fillna(df[column].mean())
```

##  **Terrorism by Years**

**terrorist incidents success from 1970 to 2018**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv(r"D:\Data Analysis\Data Analysis G.19\Python\Session\PY - DA & AI S04\Project 1\Global-Terrorism-db.csv",
                 encoding="latin-1") #utf-8
yearly_success = df.groupby('iyear')['success'].sum()
yearly_total = df.groupby('iyear').size()

plt.figure(figsize=(20,10))
plt.plot(yearly_total, label='Total Incidents', lw=3)
plt.plot(yearly_success, label='Successful Incidents', lw=3)
plt.title("Terrorist Incidents vs Successful Attacks Over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()

plt.show()
```

**Terrorist Incidents vs Successful Attacks Over Time*

```python
sns.set_theme(rc={'figure.figsize':(20, 10)}, font_scale=1.4)
plt.plot(df.groupby('iyear').sum()['nkill'],color = 'b',ls = '-',lw = 8, alpha = 0.85)           
plt.title(" Number of Kills Because of Terrorist Incidents ", size=20 )
# Add  X-axis Label 
plt.xlabel('Years',color = 'k',size=16)
# Add Y-axis Label
plt.ylabel('Number of Kills ',color = 'k',size=16)
plt.show()
```

**The number of deaths caused by terrorist incidents remained relatively low until the early 2000s, but increased significantly after 2001, reaching a sharp peak around 2014–2015. This spike corresponds to the rise of ISIS and the conflicts in Iraq and Syria. After 2015, the number of deaths gradually declined but stayed higher than pre-2000 levels.*

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
df = pd.read_csv(r"D:\Data Analysis\Data Analysis G.19\Python\Session\PY - DA & AI S04\Project 1\Global-Terrorism-db.csv",
                 encoding="latin-1") #utf-8

 
pio.renderers.default = "browser"

country_attack_counts = df['country_txt'].value_counts().reset_index()
country_attack_counts.columns = ['Country', 'Number of Attacks']

# Sort
country_attack_counts = country_attack_counts.sort_values(
    by='Number of Attacks',
    ascending=False
)

# Top 10
top10 = country_attack_counts.head(10)

# Plot
fig = px.bar(
    top10,
    x="Country",
    y="Number of Attacks",
    title="Top 10 Countries with Most Terrorist Attacks",
    text="Number of Attacks",
    color="Number of Attacks",
    color_continuous_scale="Reds"
)

fig.update_traces(textposition='outside')
fig.update_layout(xaxis_tickangle=-45)

fig.show()
```
